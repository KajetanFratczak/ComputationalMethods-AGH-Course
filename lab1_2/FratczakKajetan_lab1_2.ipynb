````julia
using CSV, Statistics, DataFrames, Plots, LinearAlgebra
n_values = [750, 1000, 1250, 1500, 1750, 2000, 3000, 4000, 5000]
output = "results.csv"

# DataFrame do przechowywania wyników
results = DataFrame(n = Int[], function_name = String[], time_elapsed = Float64[])

# Pomiary czasu 
for n in n_values
    A = rand(n, n)
    # println(typeof(A)) = Matrix{Float64} 
    x = rand(n)
    # println(typeof(x)) = Vector{Float64}
    y = rand(n)
    
    time_for_dot = Float64[]
    time_for_mult = Float64[]

    for i in 1:10
        # dot(x, A, y) = generalized dot product dot(x, A*y) (bardziej przyjazna składnia)
        time_dot = @elapsed dot(x, A, y)
        push!(time_for_dot, time_dot)
        
        time_multiplication = @elapsed A*x
        push!(time_for_mult, time_multiplication)
    end

    for t in time_for_dot
        push!(results, (n, "dot", t))
    end
    
    for t in time_for_mult
        push!(results, (n, "multiplication", t))
    end
end

CSV.write(output, results)

data = CSV.read(output, DataFrame)
````

````julia
# Pogrupowałem dane według nazwy funkcji, żeby zobaczyć różnice.
data_grouped=groupby(data, :function_name)
````

````julia
# Testowe liczenie średniej
mean(data.time_elapsed)
````

````julia
# Testowe liczenie odchylenia standardowego
std(data.time_elapsed)
````

````julia
# Grupowanie danych według funkcji i rozmiaru wektora
data_grouped = groupby(data, [:function_name, :n])

# Obliczenie średniej i odchylenia standardowego dla każdej funkcji i rozmiaru n
grouped = combine(data_grouped, :time_elapsed => mean => :mean_time, :time_elapsed => std => :std_time)

# Podział danych na dwie grupy
dot_data = filter(row -> row.function_name == "dot", grouped)
multiplication_data = filter(row -> row.function_name == "multiplication", grouped)
````

````julia
# Wykres 1
plot1 = scatter(dot_data.n, dot_data.mean_time, yerr=dot_data.std_time, seriestype=:scatter, 
    label="Pomiary", xlabel="Rozmiar macierzy (n)", ylabel="Czas wykonania (s)", title="Uogólniony iloczyn skalarny dot(x, A, y)", ylims=(0,0.045), legend=:topleft)
````

````julia
# Wykres 2
plot2 = scatter(multiplication_data.n, multiplication_data.mean_time, yerr=multiplication_data.std_time, seriestype=:scatter, 
    label="Pomiary", xlabel="Rozmiar macierzy (n)", ylabel="Czas wykonania (s)", title="Mnożenie macierzy przez wektor (A*x)", ylims=(0,0.045), legend=:topleft)
````

````julia
# Małe wyjaśnienie: 
# yerr - dodanie słupków błędów
# seriestype = :scatter - określenie rodzaju wykresu (scatter plot)
# label = "dot(x, A, y)" - opis legendy
# xlabel = "Rozmiar n" - podpis osi X
# ylabel = "Średni czas (s)" - podpis osi Y
# title = "Dot product computation" = tytuł wykresu
````

````julia
# Układ wykresów w tabelce
plot(plot1, plot2, layout=(1,2), size=(1000, 500), margin=10Plots.mm)
````
